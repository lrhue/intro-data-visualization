## Exploring Plots for Numerical Data
[Seaborn](https://seaborn.pydata.org/) provides high-level commands for plots and data exploration based on the Matplotlib packages. Although much of the functionality is the same as the  raw Matplotlib commands,  Seaborn is much more convenient.

The first step is to import the Seaborn library. Seaborn is commonly imported as *sns*.

```
import seaborn as sns
```

### Creating empty plots

Create an empty plot. Nothing will show in the output; this just changes the default settings.

There are several additional options to change the default style settings for the graphs, such as "whitegrid". 

```
sns.set()
#sns.set_style("whitegrid")
```

### Scatterplots

Let's start with scatterplots. Our first graph is a scatterplot of x and y. The basic scatterplot function is sns.scatterplot().

The first line below create the scatterplot and assign it to the variable p1. The second line uses the plt.show() function to render the saved plot.

```
p1 = sns.scatterplot(x="x", y="y", data=ex)
plt.show(p1)
```

There are multiple options to improve the look of the basic chart. 

Here is the purpose of each line below:

*   Change the color palette for Seaborn visualizations
*   Set the style of the tick marks on the x/y axis
*   Plot the scatterplot (using the same command as above)
*   Changes the labels for the x and y axis
*   Sets the limits on the x and y axes
*   Adds the title to the plot
*   Renders the plot in the output



```
sns.color_palette("Set2")
sns.set_style("ticks", {"xtick.major.size":8, "ytick.major.size":8})
p1 = sns.scatterplot(x="x", y="y", data=ex)
p1.set(xlabel="X",ylabel="Y")
p1.set(xlim=(0,60),ylim=(0,250))
plt.title("The relationship between x and y increases")
plt.show(p1)
```

Just as with the other tools, you can show multiple cues using various arguments.

In this plot, we use two additional arguments, *size* and *hue* to change the size and color of the points as a function of our third variable, z. The legend=None argument removes the legend from the chart, which is the default.

```
sns.color_palette("Set2")
p1 = sns.scatterplot(x="x", y="y",size="z", hue="z", data=ex, legend=None)
p1.set(xlabel="X",ylabel="Y")
p1.set(xlim=(0,60),ylim=(0,250))
plt.title("The relationship between x and y increases")
plt.show(p1)
```

### Regression plots

If you want to add a regression line to your plot, it is a different function in Seaborn. These are called regression plots, and the function is sns.regplot().

Here's each purpose of the lines below:

*   Plot the scatterplot with the regression line. Note that the *color* argument is "k".
*   Change the axis labels
*   Change the axis limits
*   Add the plot title
*   Show the plot


```

p2 = sns.regplot(x="x", y="y", data=ex, color="k")
p2.set(xlabel="X",ylabel="Y")
p2.set(xlim=(0,50),ylim=(0,250))
plt.title("The relationship between x and y increases")
#plt.suptitle("Another title")
plt.show(p2)
```

### Joint Plot

The joint plot shows the scatterplot of two variables in the main plot. A histogram with the distribution of each variable is on the axes.

More information about joinplot() and its arguments is available in the [documentation](http://seaborn.pydata.org/generated/seaborn.jointplot.html).



```
sns.axes_style('white')
g = sns.jointplot(x="x", y="y", data=ex)
g.ax_joint.plot(np.linspace(0, 50),
                    np.linspace(0, 250), ':k')
#g.set(xlim=(0,50),ylim=(0,250))
plt.show(g)
```

```
sns.jointplot(
    data=ex, x="x", y="y",
    marker="+", marginal_kws=dict(bins=25, fill=False),
    height=5, ratio=2, marginal_ticks=True
)
```

### Pair plot

The pair plot is useful for exploring the data. It creates a scatterplot for each pair of fields in the dataset, showing the correlations between two variable combinations. 



```
g = sns.PairGrid(ex, vars=['x', 'y', 'z'],
                 palette='RdBu_r')
g.map(plt.scatter, alpha=0.8)
g.add_legend()
```

## Exploring Charts for Time Series Data




### Line Charts

Let's practice making line charts. The basic function for line charts in Seaborn is sns.lineplot().

```
sns.set()
p3 = sns.lineplot(x="Year", y="Attrition Rate", data=attr_data_cleaned)
plt.show()
```

Year is shown as the number of years from 2010.Clearly, the default axis label for Year is insufficient. 

Python provides multiple different methods to format [the tick marks](https://matplotlib.org/stable/gallery/ticks_and_spines/tick-formatters.html). The examples below show three common needs for formatting: 

*   Showing the entire number
*   Formatting the axes as percents
*   Formatting the axes as dollars




```
p3 = sns.lineplot(x="Year", y="Attrition Rate", data=attr_data_cleaned, 
                  color="c")

xfmt = '{:g}'
xtick = mtick.FuncFormatter(lambda x, pos: xfmt.format(x))
p3.xaxis.set_major_formatter(xtick)
plt.title("Attrition Rate swings by year")

plt.show()
```

In addition, we can format the y-axis of Attrition Rate in percent. The ".0f" says to format the number in a fixed decimal of "0" after the decimal. The {}% says to add the % mark after the formatted number. In the ytick assignment, multiply the y by 100 to get the appropriate percent number.

```
p4 = sns.lineplot(x="Year", y="Attrition Rate", data=attr_data_cleaned, color="c")

xfmt = '{:g}'
tick = mtick.FuncFormatter(lambda x, pos: xfmt.format(x))
p4.xaxis.set_major_formatter(tick)

yfmt = '{:.0f}%'
ytick = mtick.FuncFormatter(lambda y, pos: yfmt.format(y*100))
p4.yaxis.set_major_formatter(ytick)

p4.yaxis.set_tick_params(which='major', labelcolor='green',
                         labelleft=True, labelright=False)

plt.show()
```

To make the y-axis in dollars, use the following format. (Note that is *not* appropriate for this chart but just for illustration.)



```
p5 = sns.lineplot(x="Year", y="Attrition Rate", data=attr_data_cleaned, color="c")

xfmt = '{:g}'
tick = mtick.FuncFormatter(lambda x, pos: xfmt.format(x))
p5.xaxis.set_major_formatter(tick)

yfmt = '${:,.2f}'
ytick = mtick.FuncFormatter(lambda y, pos: yfmt.format(y*100))
p5.yaxis.set_major_formatter(ytick)

p5.yaxis.set_tick_params(which='major', labelcolor='green',
                         labelleft=True, labelright=False)

plt.show()
```

### Bar charts

Bar charts in Seaborn do not necessarily require pre-processing the data, unlike R or Excel. Seaborn contains a [tutorial](https://seaborn.pydata.org/generated/seaborn.barplot.html) on making side-by-side bar plots using the *hue* argument. 

Seaborn does not include an easy way to make stacked bar charts, although it's possible to combine two bar charts.

```
bc1 = sns.barplot(x="Year", y="Attrition Rate", data=attr_data_cleaned, color="c")
plt.show(bc1)
```

Use the order argument to change the order of the categories.

```
attr_data_threeyears = attr_data_cleaned[attr_data_cleaned.Year >= 2017]
bc2 = sns.barplot(x="Year", y="Attrition Rate", data=attr_data_threeyears, color="b",
                  order=[2017, 2018,2019])
```
